```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# plt.rcParams["figure.figsize"] = (14,7)
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 16
```

```python
df = pd.read_csv("../data/metadata/wikipedia.csv").sort_values('avgCharCount', ascending=False, inplace=False)
df_de = pd.read_csv("../data/metadata/german-ents-wikipedia.csv").sort_values('avgCharCount', ascending=False, inplace=False)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(23, 7), dpi=300, sharey=True)
titles = ['Articles of types: human, city, public company, historical event, body of water', 'Articles of entities affiliated to Germany (e.g., German cities)']
x_label = 'Language of article'
y_data = [df.avgCharCount, df_de.avgCharCount]
x_data = [df.lang, df_de.lang]
x_ticks = [np.arange(len(df.lang)), np.arange(len(df_de.lang))]

for i in range(2):
    axs[i].set_ylim([0, int(max(y_data[i])) + 10**3])
    axs[i].set_title(titles[i], fontsize=16)
    axs[i].bar(x_data[i], y_data[i], align='center', color='black', alpha=0.75)
    if i == 0:
        axs[i].set_ylabel('Avg. number of characters per article', fontsize=16)
    axs[i].set_xlabel(x_label, fontsize=16)
    # axs[i].set_xticks(x_ticks[i], x_data[i])
    
plt.tight_layout()
plt.savefig('../data/metadata/wikipedia-stats.jpg')
```
